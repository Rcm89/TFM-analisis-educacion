### Extraemos de la web del periodico expansion los datos de Gasto publico educativo de las C.C.A.A de 2022

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://datosmacro.expansion.com/estado/presupuestos/espana-comunidades-autonomas?sc=PR-G-F-32"

# 1. Realizamos la petición GET a la página
response = requests.get(url)
response.raise_for_status()  # Lanza excepción si hay error en la petición

# 2. Parseamos el HTML con BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# 3. Localizamos la tabla por su id="tbs"
tabla = soup.find("table", id="tbs")

# 4. Obtenemos todas las filas del <tbody>
filas = tabla.find("tbody").find_all("tr")

# Listas para almacenar los datos de cada columna
ccaa_list = []
gasto_2022_list = []
porcentaje_list = []
per_capita_list = []
partido_list = []

for fila in filas:
    celdas = fila.find_all("td")

    ccaa = celdas[0].get_text(strip=True)  
    gasto_educacion = celdas[1].get("data-value")  
    porcentaje_total = celdas[3].get_text(strip=True)
    per_capita = celdas[4].get("data-value")
    partido = celdas[6].get_text(strip=True)

    # Agregamos los valores a cada lista
    ccaa_list.append(ccaa)
    gasto_2022_list.append(gasto_educacion)
    porcentaje_list.append(porcentaje_total)
    per_capita_list.append(per_capita)
    partido_list.append(partido)

# 5. Creamos el DataFrame con las columnas correspondientes
df = pd.DataFrame({
    "CCAA": ccaa_list,
    "Gasto_2022 (miles €)": gasto_2022_list,
    "% del Total": porcentaje_list,
    "Per cápita (€/persona)": per_capita_list,
    "Partido": partido_list
})

# Mostramos el DataFrame
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://datosmacro.expansion.com/estado/presupuestos/espana-comunidades-autonomas?sc=PR-G-F-32"

# 1. Realizamos la petición GET a la página
response = requests.get(url)
response.raise_for_status()  # Lanza excepción si hay error en la petición

# 2. Parseamos el HTML con BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# 3. Localizamos la tabla por su id="tbs"
tabla = soup.find("table", id="tbs")

# 4. Obtenemos todas las filas del <tbody>
filas = tabla.find("tbody").find_all("tr")

# Listas para almacenar los datos de cada columna
ccaa_list = []
gasto_2022_list = []
porcentaje_list = []
per_capita_list = []
partido_list = []

for fila in filas:
    celdas = fila.find_all("td")

    # Eliminamos "[+]" en la cadena del nombre de la CCAA
    ccaa = celdas[0].get_text(strip=True).replace("[+]", "").strip()
    gasto_educacion = celdas[1].get("data-value")
    porcentaje_total = celdas[3].get_text(strip=True)
    per_capita = celdas[4].get("data-value")
    partido = celdas[6].get_text(strip=True)

    ccaa_list.append(ccaa)
    gasto_2022_list.append(gasto_educacion)
    porcentaje_list.append(porcentaje_total)
    per_capita_list.append(per_capita)
    partido_list.append(partido)

# 5. Creamos el DataFrame con las columnas correspondientes
df = pd.DataFrame({
    "CCAA": ccaa_list,
    "Gasto_2022 (miles €)": gasto_2022_list,
    "% del Total": porcentaje_list,
    "Per cápita (€/persona)": per_capita_list,
    "Partido": partido_list
})

# Mostramos el DataFrame
df
```
